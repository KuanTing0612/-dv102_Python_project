```python
import requests
from bs4 import BeautifulSoup as bs
from lxml import html
from lxml import cssselect
```

```python
restaurant = {'thaitown':'瓦城','verythai':'非常泰','1010restaurant':'1010湘','verythainoodles':'大心','ricebar':'時時香'}
allbranch = {}
for restau in restaurant:
    res = requests.get(f"https://www.ttfb-admin.com/bs/api/v1/web/{restau}/tw/shop")
    info = res.json()
    branches = []
    for name in info["data"]:
        branch = name["name"]
        branches.append(f'{restaurant[restau]}'+branch.strip())
    allbranch[restau]=branches
allbranch
```
